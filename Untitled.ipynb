```python
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
import keras
import pandas as pd
import numpy as np
from PIL import Image
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
print(os.getcwd())
```

```python
df=pd.read_csv('cifar10Labels.csv',index_col=0)
img_index=10
print(df.label[img_index])
Image.open('cifar10/cifar10/'+str(img_index)+'.png')
```

```python
from sklearn.model_selection import train_test_split
y_train,y_test=train_test_split(df.label,test_size=0.3,random_state=42)
train_idx,test_idx=y_train.index,y_test.index
```

```python
temp=[]
for img_index in y_train.index:
    img_path=os.path.join('cifar10/cifar10/',str(img_index)+'.png')
    img=np.array(Image.open(img_path)).astype('float32')
    temp.append(img)
X_train=np.stack(temp)
```

```python
temp=[]
for img_index in y_test.index:
    img_path=os.path.join('cifar10/cifar10/',str(img_index)+'.png')
    img=np.array(Image.open(img_path)).astype('float32')
    temp.append(img)
X_test=np.stack(temp)
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
le=LabelEncoder()
X_enc=le.fit_transform(y_train)
```

```python
from keras.utils import to_categorical
y_train=to_categorical(X_enc)
```

```python
y_train
```

```python
num_classes=10
model=keras.models.Sequential([
        keras.layers.Conv2D(filters=32,kernel_size=(3,3),strides=1,padding='same',activation='relu',kernel_regularizer=keras.regularizers.l2(0.001),input_shape=(32,32,3),name='Conv_1'),
        keras.layers.BatchNormalization(name='BN_1'),keras.layers.MaxPool2D(pool_size=(2,2),name='MaxPool1'),
        keras.layers.Conv2D(filters=64,kernel_size=(3,3),strides=1,padding='same',activation='relu',kernel_regularizer=keras.regularizers.l2(0.001),input_shape=(32,32,3),name='Conv_2'),
        keras.layers.BatchNormalization(name='BN_2'),keras.layers.MaxPool2D(pool_size=(2,2),name='MaxPool2'),
        keras.layers.Flatten(name='Flat'),
        keras.layers.Dense(num_classes,activation='softmax',name='pred_layer')
    
])
```

```python
model.summary()
```

```python
# Compiling the model
model.compile(loss='categorical_crossentropy', 
              optimizer=keras.optimizers.Adam(), 
              metrics=['accuracy'])
cpfile = r'CIFAR10_checkpoint.keras'
cb_checkpoint = keras.callbacks.ModelCheckpoint(cpfile, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
epochs = 5
model.fit(X_train, y_train, epochs=epochs, validation_split=0.2, callbacks=[cb_checkpoint])
```
